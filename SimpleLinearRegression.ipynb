```python
#importing Libraries
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inLine
import pandas as pd
```

```python
#Importing datasets
dataset = pd.read_csv('Salary_Data.csv')
```

```python
dataset
```

```python
#Definning your matrix of Features and Dependent variable
X = dataset.iloc[:,:1]
y = dataset.iloc[:, 1]
```

```python
y
```

```python
X
```

```python
#Fitting your regressors
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(X,y)
```

```python
#Ploting your chart
plt.scatter(X,y, color="red")
plt.plot(X, regressor.predict(X), label="Predict")
plt.title('our 1st Linear regression chart')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.legend(loc=2)
plt.show()
plt.close()
```
